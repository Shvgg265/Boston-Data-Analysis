```python
from sklearn import datasets
import numpy as np

# linear_model is used for linear regression
from sklearn import linear_model
from sklearn.utils import shuffle
import pandas as pd

# For heatmaps
import seaborn as sns
import matplotlib.pyplot as plt

# For Train-Test
from sklearn.model_selection import train_test_split
```

```python
# Loaded the in-built dataset boston 
data = datasets.load_boston()
```

```python
# Typecasting it to DataFrame
boston = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
# Adding new column
boston['TARGET'] = data.target
```

```python
X = boston[['INDUS', 'RM', 'TAX', 'PTRATIO', 'LSTAT']] 
```

```python
y = boston['TARGET']
y
```

```python
# Cross Validation - Split data into X and Y
# It'll allocate memory randomly
# Avoid overfitting - 
# random_state is similar to np.seed. It freezes the split.
X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

## Linear Regression

```python
lrmodel = linear_model.LinearRegression()
lrmodel.fit(X_train, Y_train)
```

```python
y_predict = lrmodel.predict(X_train)
```

```python
lrmodel.score(X_train,Y_train)
```

```python
lrmodel.score(X_test,Y_test)
```

```python
# y = beta0 + beta1(x1) + beta2(x2) + ...

# beta0
lrmodel.intercept_
```
